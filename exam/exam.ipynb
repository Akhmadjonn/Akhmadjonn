# Decision Treedan foydalanamiz

```python
# model train qilamiz

model = DecisionTreeClassifier(random_state=42)
model.fit(x_train, y_train)
```

```python
# Predicting
y_pred = model.predict(x_test)

y_pred
```

```python
# K-fold cross-validation (k=5)
cv_scores = cross_val_score(model, x_scaled, y, cv=5)  # cv=5 means 5-fold cross-validation
avg_cv_scores=np.mean(cv_scores)

#O'rtacha qiymat topamiz
avg_cv_scores
```
