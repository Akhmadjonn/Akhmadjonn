```python
cards_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Project15/cards_data.csv')
users_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Project15/users_data.csv')
transactions_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Project15/transactions_data.csv')
```

```python

merged_df = pd.merge(users_df, transactions_df, left_on='id', right_on='client_id', how='inner')

```

```python
merged_df.info()
```

```python
merged_df.drop(['num_credit_cards', 'id_x', 'id_y', 'client_id', 'errors'], axis=1, inplace=True)
```

```python
merged_df.info()
```

```python
merged_df = pd.merge(merged_df, cards_df, left_on="card_id", right_on="id", how="inner")
```

```python
merged_df.info()
```

```python
drop_cols = ['birth_year', 'birth_month','address', 'date', 'card_id', 'use_chip', 'merchant_state',
             'zip', 'id', 'client_id', 'card_number', 'cvv', 'merchant_id', 'expires',
             'num_cards_issued', 'year_pin_last_changed', 'card_on_dark_web', 'retirement_age']

merged_df.drop(drop_cols, axis=1, inplace=True)
```

```python
merged_df.info()
```

```python
merged_df['acct_open_date'] = pd.to_datetime(merged_df['acct_open_date'])

current_date = datetime.now()

merged_df['account_age'] = (current_date - merged_df['acct_open_date']).dt.days // 365

merged_df.drop('acct_open_date', axis=1, inplace=True)
```

```python
merged_df.info()
```

```python
merged_df.head()
```

```python
merged_df['card_brand'].value_counts()
```

```python
merged_df.isnull().sum()
```

```python
num_col = ['per_capita_income', 'yearly_income', 'total_debt', 'amount', 'credit_limit']
```

```python
for col in num_col:
  merged_df[col] = merged_df[col].astype(str).str.replace(r'[$,]', '', regex=True)
  merged_df[col] = merged_df[col].replace('', pd.NA)
  merged_df[col] = pd.to_numeric(merged_df[col], errors='coerce')

merged_df.info()
```

```python
merged_df.isnull().sum()
```

# spliting


```python
x = merged_df.drop('card_brand', axis=1)
y = merged_df['card_brand']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
train = pd.concat([x_train, y_train], axis=1)
test = pd.concat([x_test, y_test], axis=1)

train.to_csv("train.csv", index=False)
test.to_csv("test.csv", index=False)
```

```python
group_A = merged_df[merged_df['card_brand'] == 'Mastercard']

delete_A = int(len(group_A) * 0.8)

rows_delete_A = group_A.sample(n=delete_A, random_state=42)

df_cleaned = merged_df.drop(rows_delete_A.index).reset_index(drop=True)
```

```python
group_B = df_cleaned[df_cleaned['card_brand'] == 'Visa']
delete_B = int(len(group_B) * 0.7)
rows_delete_B = group_B.sample(n=delete_B, random_state=42)
df_short = df_cleaned.drop(rows_delete_B.index).reset_index(drop=True)
```
