```python
#Categorical-> Numerical
categorical_col=df.select_dtypes(include=['object','category']).columns
categorical_col
```

```python
#One -hot encoding(dataset kichik bolganda)

df_one_hot=pd.get_dummies(df, columns=categorical_col, drop_first=True)
df_one_hot.head()
```

```python
# Separate numeric columns (excluding binary one-hot encoded columns)
numeric_columns = df_one_hot.select_dtypes(include=['float64', 'int64']).columns

# Fill missing values only in numeric columns with their respective mean
df_one_hot[numeric_columns] = df_one_hot[numeric_columns].fillna(df_one_hot[numeric_columns].mean())


# df_one_hot[categorical_columns] = df_one_hot[categorical_columns].fillna(0)

df_one_hot
```

```python
#1. 1 dan 10 gacha bolgan juft sonlar ni toping
a = 1
while a<=10:
    if a%2==0:
        print(a, end=" ")
    a +=1
```

```python
#2. 1 dan 10 gacha bo'lgan toq sonlarni toping
a = 1
while a<=10:
    if a%2==1:
        print(a, end=" ")
    a +=1
```

```python
#3. Yuqoridagi xolatlarda yigiindi toping
a = 1
sum1, sum2 = 0, 0
while a<=10:
    if a%2==0:
        sum1 += a
    else:
        sum2 += a
    a +=1
print("Toq sonlar yig'indisi: ", sum2)
print("Juft sonlar yig'indisi: ", sum1)
```

```python
#4 Yuqoridagi xolatlarda ko'paytmani toping
a = 1
sum1, sum2 = 1, 1
while a<=10:
    if a%2==0:
        sum1 *= a
    else:
        sum2 *= a
    a +=1
print("Toq sonlar ko'paytmasi: ", sum2)
print("Juft sonlar ko'paytmasi: ", sum1)
```

```python
#5. 1 dan 100 gacha sonlar ichida 3 ga bo'linadigan sonlarni yigindisini toping
a = 3
sum1 = 0
while a <= 100:
    sum1 += a
    a += 3
print(sum1)
```
